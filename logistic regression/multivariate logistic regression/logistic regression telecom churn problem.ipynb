```python
import pandas as pd
```

```python
churn_data=pd.read_csv("churn_data.csv")
customer_data=pd.read_csv("customer_data.csv")
internet_data=pd.read_csv("internet_data.csv")
```

```python
churn_data.head()
```

```python
customer_data.head()
```

```python
internet_data.head()
```

```python
df=pd.merge(churn_data,customer_data,how="inner",on="customerID")
```

```python
telecom_data=pd.merge(df,internet_data,how="inner",on="customerID")
telecom_data.head()
```

```python
telecom_data.info()
```

```python
telecom_data.PhoneService=telecom_data.PhoneService.map({"Yes":1,"No":0})
telecom_data.PaperlessBilling=telecom_data.PaperlessBilling.map({"Yes":1,"No":0})
telecom_data.Churn=telecom_data.Churn.map({"Yes":1,"No":0})
telecom_data.Partner=telecom_data.Partner.map({"Yes":1,"No":0})
telecom_data.Dependents=telecom_data.Dependents.map({"Yes":1,"No":0})
```

```python
contact=pd.get_dummies(telecom_data["Contract"],prefix="Contract",drop_first=True)
telecom_data=pd.concat([telecom_data,contact],axis=1)


pay=pd.get_dummies(telecom_data.PaymentMethod,prefix="PaymentMethod",drop_first=True)
telecom_data=pd.concat([telecom_data,pay],axis=1)

gen=pd.get_dummies(telecom_data.gender,prefix="gender",drop_first=True)
telecom_data=pd.concat([telecom_data,gen],axis=1)

mul_lines=pd.get_dummies(telecom_data.MultipleLines,prefix="MultipleLines",drop_first=True)
telecom_data=pd.concat([telecom_data,mul_lines],axis=1)

internet_service=pd.get_dummies(telecom_data.InternetService,prefix="InternetService",drop_first=True)
telecom_data=pd.concat([telecom_data,internet_service],axis=1)

online_security=pd.get_dummies(telecom_data.OnlineSecurity,prefix="OnlineSecurity",drop_first=True)
telecom_data=pd.concat([telecom_data,online_security],axis=1)

online_backup=pd.get_dummies(telecom_data.OnlineBackup,prefix="OnlineBackup",drop_first=True)
telecom_data=pd.concat([telecom_data,online_backup],axis=1)


device_protection=pd.get_dummies(telecom_data.DeviceProtection,prefix="DeviceProtection",drop_first=True)
telecom_data=pd.concat([telecom_data,device_protection],axis=1)

tech_support=pd.get_dummies(telecom_data.TechSupport,prefix="TechSupport",drop_first=True)
telecom_data=pd.concat([telecom_data,tech_support],axis=1)


streaming_tv=pd.get_dummies(telecom_data.StreamingTV,prefix="StreamingTV",drop_first=True)
telecom_data=pd.concat([telecom_data,streaming_tv],axis=1)


streaming_movies=pd.get_dummies(telecom_data.StreamingMovies,prefix="StreamingMovies",drop_first=True)
telecom_data=pd.concat([telecom_data,streaming_movies],axis=1)

```

```python
telecom_data=telecom_data.drop(["Contract","PaymentMethod","gender",
                               "MultipleLines","InternetService","OnlineSecurity",
                               "OnlineBackup","DeviceProtection","TechSupport",
                                "StreamingTV","StreamingMovies"],1)
```

```python
telecom_data.TotalCharges=telecom_data.TotalCharges.convert_objects(convert_numeric=True)
telecom_data.info()
```

```python
# standarization

df=telecom_data[["tenure","MonthlyCharges","TotalCharges"]]
```

```python
standard_df=(df-df.mean())/df.std()
```

```python
telecom_data=telecom_data.drop(["tenure","MonthlyCharges","TotalCharges"],1)
```

```python
telecom_data=pd.concat([telecom_data,standard_df],1)
```
