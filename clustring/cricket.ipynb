```python
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from sklearn.cluster import KMeans
import seaborn as sns
from datetime import datetime
```

```python
cricket = pd.read_csv("Cricket.csv",  sep = ',',encoding = "ISO-8859-1", header= 0)
```

```python
data=cricket[["Mat","Inns","NO","Runs","Ave","BF","SR","100","50","0"]]
from sklearn.preprocessing import StandardScaler
standard_scaler = StandardScaler()
data = standard_scaler.fit_transform(data)
```

```python
data
model_clus5 = KMeans(n_clusters = 4, max_iter=50)
model_clus5.fit(data)
```

```python
cricket.index = pd.RangeIndex(len(cricket.index))
x=list(cricket.columns)
x=[i for i in x]
x.append("c_id")
cricket = pd.concat([cricket, pd.Series(model_clus5.labels_)], axis=1)
cricket.columns = x
cricket
```
